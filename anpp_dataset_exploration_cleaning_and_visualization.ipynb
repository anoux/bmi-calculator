```python
my_list = [2547.658114, 9012.541957, 6402.379562, 15254.226219]
my_index = ['Grassland', 'Pasture', 'Winter crop', 'Summer crop']


my_series = pd.Series(my_list, index=my_index)
```

```python
df_by_resource_type = pd.DataFrame({'Resource': my_series.index, 'mean ANPP': my_series.values})
```

```python
df_by_resource_type
```

```python
df_by_resource_type.sort_values(by="mean ANPP",ascending=False)

# Summer short term crops are the most productive ones, then pastures,
# the winter short term crops, and last of all grasslands
```

```python
df_by_resource_type = df_by_resource_type.reset_index(drop=True)

df_by_resource_type
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab


fig_df_by_resource_type = df_by_resource_type.set_index('Resource')
color_bars = ['green', 'yellow', 'blue', 'orange']

# So the names for each colmun along the x axis are 
#the different resources instead on numbers

fig_df_by_resource_type.plot(kind = 'bar', title = 'Mean ANPP by resource',
                        xlabel = 'Resource', 
                        ylabel = 'Mean ANPP (kg/ha.year-1)',
                        rot=0,
                        figsize = (16,8),
                        legend = False)


#Font size setting for plots
 
SMALL_SIZE = 15
MEDIUM_SIZE = 20
BIGGER_SIZE = 30

plt.rc('font', size=MEDIUM_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)    # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=MEDIUM_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
```
